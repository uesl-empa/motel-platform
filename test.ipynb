```python
import ollama
from pydantic import BaseModel, Field
from typing import Literal

# Define your classification schema
class DataClassification(BaseModel):
    category: Literal["spam", "ham", "news", "complaint", "support", "other"]
    confidence: float = Field(..., ge=0, le=1)
    reason: str
    tags: list[str]

response = ollama.chat(
    model='qwen3:14b',
    messages=[{
        'role': 'user',
        'content': 'Classify the following text...'
    }],
    format=DataClassification.model_json_schema()   # ← This enforces the schema
)

# Parse directly
result = DataClassification.model_validate_json(response['message']['content'])
print(result)
```
